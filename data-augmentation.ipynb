# Generamos imagenes rotadas + aplicamos brillo

```python
import PIL, os
from PIL import Image, ImageEnhance
import random

# change to directory where image is located
dir_path = r'C:\Users\Miguel Angel\Desktop\dataset\img'
input_prefix = 'img'

os.chdir(dir_path) 
height= 512
width= 512
rotate =[0,90,180,270]

images_length = 350

i=1
n=1
while i < images_length+1:
    j=0
    while j < len(rotate):
        
        #Load image
        input_image = "{} ({}).png".format(input_prefix, i)
        picture= Image.open(input_image)
        #Rotate
        picture = picture.rotate(rotate[j])  
        #Add random brightness
        enhancer = ImageEnhance.Brightness(picture)
        output_image = enhancer.enhance(round(random.uniform(0.3, 2), 2))   
        #Save output image
        output_prefix = "{}x{} ({}).png".format(height,width, n)      
        output_image.save(output_prefix)
        j += 1  
        n += 1
        print(output_prefix)
       
    i += 1

```

# Distribuimos nuestro dataset en subconjuntos train/validation/test
